```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import matplotlib.cm as cm
import re
from ipywidgets import interact, widgets, fixed

dxmin=1.0e-9
dxmax=1.0e-2
dymin=0
dymax=12

format_string = lambda s: "INP = {}, FLARE = {}".format(*re.findall(r'[0-9\.e]+', s), 'FEno' if 'FEno' in s else '') 

class MultiPanelPlot:
    def __init__(self, datasets, varname, vmin=1.0, vmax=1.0e4):
        self.datasets, self.varname = datasets, varname
        key = next(iter(datasets))
        self.time = datasets[key].time.values
        self.x = datasets[key].x.values
        self.y = datasets[key].y.values
        self.height = HHLd
        self.bins = np.arange(datasets[key][varname].shape[-1])
        self.rgrenz = RGRENZ

        self.standard_name, self.unit = datasets[key][varname].attrs.values()
        self.cmap = mpl.colors.LinearSegmentedColormap.from_list('mycmap', [(0, 'white'), (0.2, 'blue'), (0.4, 'cyan'), (0.5, 'lime'), (0.6, 'yellow'), (0.9, 'red'), (1, 'purple')]) 
        self.norm = colors.LogNorm(vmin, vmax)
    
    def display(self, timestep=0, x=12, y=12, savefig=False, xmin=dxmin, xmax=dxmax, ymin=dymin, ymax=dymax):
        mpl.rcParams['font.size'] = 13
        fig, axes = plt.subplots(3, 4, figsize=(25, 15))
        self.axes = axes.flatten()

        for i, (key, ds) in enumerate(self.datasets.items()):
            if self.varname in ds:
                z = ds[self.varname][timestep, :, y, x, :]
                self.axes[i].pcolormesh(self.rgrenz, self.height, z, cmap=self.cmap, norm=self.norm)
                self.axes[i].set(title=format_string(key), ylim=[ymin, ymax], xlim=[xmin, xmax], xscale='log')
                self.axes[i].tick_params(which='both', direction='in')
                self.axes[i].minorticks_on()
                self.axes[i].grid(True, which='major', linestyle='-', linewidth='0.5', color='black', alpha=0.5)
                self.axes[i].grid(True, which='minor', linestyle=':', linewidth='0.5', color='black', alpha=0.25)

        for i, label in enumerate(['high', 'mid', 'low']):
            ax = self.axes[i*4+3]
            ax.yaxis.set_label_position("right")
            ax.set_ylabel(label)

        fig.subplots_adjust(bottom=0.2)  # make space for colorbar
        self.cbar_ax = fig.add_axes([0.15, 0.15, 0.7, 0.02])
        self.cbar = fig.colorbar(cm.ScalarMappable(norm=self.norm, cmap=self.cmap), cax=self.cbar_ax, orientation='horizontal', extend='both')
        self.cbar.set_label(f'{self.standard_name} [{self.unit}]', size=14)
        fig.subplots_adjust(hspace=0.3, wspace=0.15, bottom=0.2)

        plt.suptitle(f"COSMO-SPECS -- Eriswil -- {self.time[timestep].astype('datetime64[s]').tolist().strftime('%Y-%m-%d_%H:%M:%S [UTC]')}", fontsize=16)
        
        if savefig:
            fig.savefig(f"{self.time[timestep].astype('datetime64[s]').tolist().strftime('%Y%m%d_%H%M%S')}_{self.varname}.png", dpi=300, facecolor='white')

    def interactive(self):
        interact(
            self.display, 
            timestep=widgets.IntSlider(min=0, max=len(self.time)-1, step=1, value=0), 
            x=widgets.IntSlider(min=0, max=len(self.x)-1, step=1, value=12), 
            y=widgets.IntSlider(min=0, max=len(self.y)-1, step=1, value=12), 
            xmin=fixed(dxmin), xmax=fixed(dxmax), ymin=fixed(dymin), ymax=fixed(dymax),
            savefig=fixed(False)
            )
```

```python
import concurrent.futures
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import matplotlib.cm as cm
import re
from ipywidgets import interact, widgets, fixed

dxmin=1.0e-9
dxmax=1.0e-2
dymin=0
dymax=12

format_string = lambda s: "INP = {}, FLARE = {}".format(*re.findall(r'[0-9\.e]+', s), 'FEno' if 'FEno' in s else '') 

class MultiPanelPlot:
    def __init__(self, datasets, varname, vmin=1.0, vmax=1.0e4):
        self.datasets, self.varname = datasets, varname
        key = next(iter(datasets))
        self.time = datasets[key].time.values
        self.x = datasets[key].x.values
        self.y = datasets[key].y.values
        self.height = HHLd
        self.bins = np.arange(datasets[key][varname].shape[-1])
        self.rgrenz = RGRENZ

        self.standard_name, self.unit = datasets[key][varname].attrs.values()
        self.cmap = mpl.colors.LinearSegmentedColormap.from_list('mycmap', [(0, 'white'), (0.2, 'blue'), (0.4, 'cyan'), (0.5, 'lime'), (0.6, 'yellow'), (0.9, 'red'), (1, 'purple')]) 
        self.norm = colors.LogNorm(vmin, vmax)

        mpl.rcParams['font.size'] = 10
        self.fig, self.axes = plt.subplots(3, 4, figsize=(15, 10))
        self.axes = self.axes.flatten()

        self.pmeshs = []
        for i, (key, ds) in enumerate(self.datasets.items()):
            if self.varname in ds:
                z = ds[self.varname][0, :, 12, 12, :]
                pmesh = self.axes[i].pcolormesh(self.rgrenz, self.height, z, cmap=self.cmap, norm=self.norm)
                self.pmeshs.append(pmesh)
                self.axes[i].set(title=format_string(key), ylim=[dymin, dymax], xlim=[dxmin, dxmax], xscale='log')
                self.axes[i].tick_params(which='both', direction='in')
                self.axes[i].minorticks_on()
                self.axes[i].grid(True, which='major', linestyle='-', linewidth='0.5', color='black', alpha=0.5)
                self.axes[i].grid(True, which='minor', linestyle=':', linewidth='0.5', color='black', alpha=0.25)
        
        self.fig.subplots_adjust(bottom=0.2)  # make space for colorbar
        self.cbar_ax = self.fig.add_axes([0.15, 0.15, 0.7, 0.02])
        self.cbar = self.fig.colorbar(cm.ScalarMappable(norm=self.norm, cmap=self.cmap), cax=self.cbar_ax, orientation='horizontal', extend='both')
        self.cbar.set_label(f'{self.standard_name} [{self.unit}]', size=14)
        self.fig.subplots_adjust(hspace=0.3, wspace=0.15, bottom=0.2)
    
    def display(self, timestep=0, x=12, y=12, savefig=False):
        for i, (key, ds) in enumerate(self.datasets.items()):
            if self.varname in ds:
                z = ds[self.varname][timestep, :, y, x, :].values
                self.pmeshs[i].set_array(z)

        plt.suptitle(f"COSMO-SPECS -- Eriswil -- {self.time[timestep].astype('datetime64[s]').tolist().strftime('%Y-%m-%d_%H:%M:%S [UTC]')}", fontsize=16)
        plt.show()
        
        if savefig:
            self.fig.savefig(f"{self.time[timestep].astype('datetime64[s]').tolist().strftime('%Y%m%d_%H%M%S')}_x{x}_y{y}_{self.varname}.png", dpi=300, facecolor='white')

    def interactive(self):
        interact(
            self.display, 
            timestep=widgets.IntSlider(min=0, max=len(self.time)-1, step=1, value=0), 
            x=widgets.IntSlider(min=0, max=len(self.x)-1, step=1, value=12), 
            y=widgets.IntSlider(min=0, max=len(self.y)-1, step=1, value=12), 
            savefig=fixed(False)
            )
        
    def save_figure(self, timestep, x=12, y=12):
        self.display(timestep=timestep, x=x, y=y, savefig=True)

    def save_all_figures(self, x=12, y=12):
        with concurrent.futures.ProcessPoolExecutor() as executor:
            executor.map(self.save_figure, range(len(self.time)), [x]*len(self.time), [y]*len(self.time))
```

### Plotter Functions

```python
def specs_time_plotter( specs, vname, vslice =  slice(8000,1000,2), do_log = True, **kwargs ):

    
    v = specs[vname].sel(HMLd = vslice)
    if do_log:
        v = np.log( v ) 
        
    
    g = v.plot(x = 'bins', y = 'time', col = 'HMLd', col_wrap = 4, 
                                                            vmin = -30)
    plt.xscale('log')
    plt.gca().invert_yaxis()

    for i, ax in enumerate(g.axes.flat):
        try:
            h = 1e-3 * v['HMLd'].isel( HMLd = i )
            ax.set_title(ts.strftime('%.1f km' % h.data) )

        except:
            pass

    for a in g.axes[-1]:
        a.set_xlabel('radius / m')
        
    return g
```

```python
def specs_height_plotter( specs, vname, tslice =  slice(0,None,10), do_log = True, save_fig = True, **kwargs ):

    pdict = dict()
    
    v = specs[vname]
    if do_log:
        v = np.log( v )
        pdict.update( dict(vmin = -30) )
    else:
        vabs = np.abs( v.data )
        ma = np.isfinite( vabs ) & (vabs != 0)
        
        if np.sum( ma ) != 0:
            v10 = np.percentile(vabs[ma], 10)
            pdict.update( dict( norm = plt.matplotlib.colors.SymLogNorm( linthresh = v10)) )
        
    g = v.isel(time = tslice).plot(x = 'bins', col = 'time', y = 'HMLd', robust = True,
                                                           col_wrap = 8, **pdict)

    for i, ax in enumerate(g.axes.flat):
        try:
            t = specs.isel(time = tslice).isel(time=i)
            ts = pd.to_datetime(str(t.time.data)) 
            ax.set_title(ts.strftime('%H:%M'))

        except:
            pass

    plt.xscale('log')
    plt.ylim(0,10e3)

    for a in g.axes[-1]:
        a.set_xlabel('radius / m')
        
    if save_fig:
        plt.savefig('pics/bin_height_plot_%s.jpg' % vname ,  )
        plt.close()
        
    return g
```
